```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv("data/students.csv")
df = df.drop(df[df.Target=='Enrolled'].index)
df.head()
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Target",data=df,ax=ax)
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Gender",data=df, hue="Target",multiple="dodge",discrete=True,shrink=.8)
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Age at enrollment",data=df, hue="Target")
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Target",data=df, hue="Scholarship holder",multiple="dodge",discrete=True,shrink=.8,legend=True)
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Educational special needs",data=df, hue="Target",multiple="dodge",discrete=True,shrink=.8,legend=True)
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Tuition fees up to date",data=df, hue="Target",multiple="dodge",discrete=True,shrink=.8,legend=True)
```

```python
fig,ax = plt.subplots()
sns.histplot(x="Unemployment rate",data=df, hue="Target",multiple="stack",discrete=False,shrink=.8,legend=True)
```
